```python
import pandas as pd
df = pd.read_csv("/content/Large Cars Dataset.csv")
df
```

```python
# There are 2 columns(MSRP and DealerCost) which has special charater("$") and comma so removing them
df["MSRP"] = df["MSRP"].str.replace("$","")
df["MSRP"] = df["MSRP"].str.replace(",","")
df["DealerCost"] = df["DealerCost"].str.replace("$","")
df["DealerCost"] = df["DealerCost"].str.replace(",","")
df
```

```python
# Loop through columns with object dtype and print their names along with their number of unique values
for i in df.select_dtypes(include=["object"]).columns:
  print(i,df[i].nunique())
```

```python
# type casting MSRP and DealerCost columns, since they are in numerical form
df['MSRP'] = df['MSRP'].astype('Int64')
df['DealerCost'] = df['DealerCost'].astype('Int64')
for i in df.select_dtypes(include=["object"]).columns:
  print(i,df[i].nunique())
```

```python
# Checking for null values
df.isnull().sum()
```

```python
# Extracting the column contains null value
df[df['Cylinders'].isnull()]
```

```python
# As per domain reasearch, Sports car with HP > 195 should have 4 cylinders for smoothness and balance. Decided to fill 4 cylinders in nulls
df.loc[df["Cylinders"].isnull(),"Cylinders"]=df[df["Cylinders"].isnull()].apply(lambda i: 2 if ((i["VehicleClass"] == 'Sports') & (i["HorsePower"] > 195) & (i["EngineSize"] == 1.3)) else 6,axis=1)
```

```python
# Converting objects/ string into meaningful numerical value(encoding)
from sklearn.preprocessing import OrdinalEncoder
encode = OrdinalEncoder()
df["Brand"] = encode.fit_transform(df[["Brand"]])
df["Model"] = encode.fit_transform(df[["Model"]])
df["VehicleClass"] = encode.fit_transform(df[["VehicleClass"]])
df["Region"] = encode.fit_transform(df[["Region"]])
df["DriveTrain"] = encode.fit_transform(df[["DriveTrain"]])
df
```

```python
# Seperated the continuous and categories columns
continuous =['Model', 'MSRP', 'DealerCost', 'EngineSize', 'Cylinders',
             'HorsePower', 'MPG_City', 'MPG_Highway', 'Weight', 'Wheelbase', 'Length']

categories=['Brand', 'VehicleClass', 'Region','DriveTrain']
```

```python
from scipy import stats # Using package of statistic
```

```python
#continious vs continious columns are compared with two sample t test
def two_sample(d1,d2):
  t=0
  f=0
  for i in  range(31):
    sample1=d1.sample(frac=0.03)
    sample2=d2.sample(frac=0.03)
    t_test,p_value=stats.ttest_ind(sample1,sample2)
    if p_value < 0.05:
      f=f+1
    else:
      t=t+1
  if t>f:
    return True
  else:
    return False
```

```python
two_sample(df["Weight"],df["Wheelbase"])
```

```python
#category vs category columns are compared with chi square test
def chisquare_cat_vs_cat(d1,d2):
    return True if stats.chi2_contingency(pd.crosstab(d1,d2))[1] < 0.05 else False
```

```python
chisquare_cat_vs_cat(df["Region"],df["DriveTrain"])
```

```python
#continious(d1) vs category(d2) columns are compared with ANNOVA test
def annova_test(d1,d2):
    group= df[d2].unique()
    data = {}
    for i in group:
        data[i]=df[d1][df[d2]==i]
    f_value,p_value=stats.f_oneway(*[i for i in data.values()])
    return False if p_value < 0.05 else True
```

```python
annova_test("Model","Brand")
```

```python
# Comparing all columns within dataframe using two sample t, chisquare and ANNOVA test
final = {}
for i in df.columns:
    final[i]={}
    for j in df.columns:
        if (i in continuous) and (j in continuous):
            result=two_sample(df[i],df[j])
        elif (i in continuous) and (j in categories):
            result=annova_test(i,j)
        elif (i in categories) and (j in continuous):
            result=annova_test(j,i)
        elif (i in categories) and (j in categories):
            result=chisquare_cat_vs_cat(df[i],df[j])
        if result:
            final[i][j]=1
        else:
            final[i][j]=0
```

```python
# Printing the values of compared results
import pprint
pprint.pprint(final)
```

```python
# Displaying the compared data into data frame
final_df = pd.DataFrame(final)
final_df
```
